```python
def make_corpus(sample_text):
  sentence_level_slice = sample_text.replace('\n', '. ').split('. ')
  sentence_level_slice = [sentence for sentence in sentence_level_slice if sentence != '']  #?? meaning

  word_level_slice = []
  for sentence in sentence_level_slice:
    sentence_in_word = []
    for word in sentence.split(' '):
      sentence_in_word.append(word.lower()) #word turns to lowercase and save to sentence_in_word
    word_level_slice.append(sentence_in_word) #?? why creating word_level_slice when sentence_level_slice is there to use
    print(sentence_in_word)
    print(word_level_slice)
  return word_level_slice
corpus = make_corpus(sample_text)
```
